```python
import glob
import random
import os

random.seed(42)
DEST_FOLDER = "/home/nv3le/workspace/Doping/run_benchmark/all_dataset/"
# all_chc_files = glob.glob("/home/nv3le/workspace/Doping/all_lra_chc18/*.smt2")
# all_chc_files = sorted(all_chc_files)

with open("lra-mutable-files", "r") as f:
    all_chc_files = f.readlines()

N_MUTATION = 10
# print(all_chc_files)

def is_real(s):
    return s.replace('.','',1).isdigit()

def mutate_file(filename, new_filename, perm_range = 0.05):
    print(filename)
    with open(filename, "r") as f:
        raw = f.readlines()
    
    #find the property assert
    last_assert_index = -1
    for i, l in enumerate(raw):
        if l.strip()=="(assert":
            last_assert_index = i
    print(last_assert_index)
    n_val = 0
    #randomly mutate constant in subsequent lines
    for i in range(last_assert_index, len(raw)):
        line = raw[i]
        trailing_spaces = len(line) - len(line.lstrip())
        tokens = line.split()
        for token_idx, t in enumerate(tokens):
            if is_real(t):
                val = float(t)
                if val!=0:
                    n_val+=1
                new_val = random.uniform(val*(1-perm_range), val*(1+perm_range))
                tokens[token_idx] = "%.2f" % new_val
            elif is_real(t[:-1]):
                
                val = float(t[:-1])
                if val!=0:
                    n_val+=1
                new_val = random.uniform(val*(1-perm_range), val*(1+perm_range))
                tokens[token_idx] = "%.2f)" % new_val
        raw[i] = " "*trailing_spaces+" ".join(tokens)+"\n"
        
    if n_val >0:
        if not os.path.exists(os.path.dirname(new_filename)):
            os.makedirs(os.path.dirname(new_filename))
        with open(new_filename, "w") as f:
            f.writelines(raw)
    return new_filename, n_val
    
mutable_files = []
variant_files = []
for chc_file in all_chc_files:
    chc_file = chc_file.strip()
    base_filename = os.path.basename(chc_file)
    for i in range(N_MUTATION):
        mutable_files.append(chc_file+"\n")
        new_filename, n_val = mutate_file(chc_file, 
                            os.path.join(DEST_FOLDER, base_filename+".folder", "variant_{}".format(i), "input_file.smt2")
                           )
        variant_files.append(new_filename+"\n")
with open("lra-variant-files", "w") as f:
    f.writelines(variant_files)
    
# with open("lra-mutable-files", "w") as f:
#     f.writelines(mutable_files)
        
```

```python
with open("lra-solved-files", "r") as f:
    solved_files = f.readlines()
solved_mutable_files = set(solved_files).intersection(set(mutable_files))
unsolved_mutable_files = set(mutable_files).difference(solved_mutable_files)

solved_mutable_files = sorted(list(solved_mutable_files))

solved_mutable_count = {}
for f in solved_mutable_files:
    base = os.path.basename(f.strip())
    try:
        ind_gen_path = os.path.join(DEST_FOLDER, "{}.folder".format(base), "ind_gen_files")
        no_of_files = len(os.listdir(ind_gen_path))
        solved_mutable_count[f] = no_of_files
        print(f, no_of_files)
    except:
        solved_mutable_count[f] = 0

solved_mutable_files_sorted = [k for k, v in sorted(solved_mutable_count.items(), key=lambda item: item[1])]
print(solved_mutable_count)
unsolved_mutable_files = sorted(list(unsolved_mutable_files))
with open("lra-unsolved-mutable-files", "w") as f:
    f.writelines(unsolved_mutable_files)
with open("lra-solved-mutable-files", "w") as f:
    f.writelines(solved_mutable_files_sorted)
```
