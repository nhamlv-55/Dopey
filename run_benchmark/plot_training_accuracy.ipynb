```python
import matplotlib
import glob
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

now = datetime.now() # current date and time
time = now.strftime("%m-%d-%H-%M-%S")
def plot_one(axes, report_name):
    res = {"accurarcy": [],
          "f1": [],
          "precision": [],
          "recall": []}
    
    with open(report_name, "r") as f:
        raw = f.readlines()
        for l in raw:
#             print(l)
            for k in res:
                if l.startswith(k):
                    tokens = l.strip().split()
                    value = float(tokens[-1])
                    res[k].append(value)
#     print(res)
    if len(res["f1"])>=30:
        axes[0].plot(res["accurarcy"][-30:])
        axes[1].plot(res["f1"][-30:])
        axes[2].plot(res["precision"][-30:])
        axes[3].plot(res["recall"][-30:])
        print(report_name)
        if res["f1"][-1]<0.7:
            print("f1 = {} < 0.7: {}".format(res["f1"][-1], report_name))
    return res

def plot_all(report_name):
    fig, axes = plt.subplots(4, figsize=(10, 40))
    axes[0].set_title("Acc")
    axes[1].set_title("F1")
    axes[2].set_title("Precision")
    axes[3].set_title("Recall")
    training_reports = glob.glob("all_dataset/*/{}".format(report_name))
#     print(training_reports)
    for report in training_reports:
        
        plot_one(axes, report)
    
    fig.show()
    fig.savefig('plot-{}-{}.pdf'.format(report_name, datetime))
```

```python
plot_all("X_train_stdout_pos")
```

```python
plot_all("X_train_stdout_neg")
```
