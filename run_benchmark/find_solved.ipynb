```python
'''
Find chc files those are solved by z3
'''
import glob
import os
all_smt2_file = glob.glob("/home/nv3le/workspace/Doping/all_lra_chc18/*.smt2")
all_smt2_file = sorted(all_smt2_file)
solved = []
for s in all_smt2_file:
    base = os.path.basename(s)
    print(base)
    with open("/home/nv3le/workspace/Doping/run_benchmark/all_dataset/{}.folder/stdout".format(base),
              "r") as f:
        raw = f.readlines()
        if len(raw)==0: continue
        if "timeout" in raw[-1]:
            print(s)
        else:
            solved.append(s+"\n")
with open("/home/nv3le/workspace/Doping/run_benchmark/lra-solved-files", "w") as f:
    f.writelines(solved)
```
